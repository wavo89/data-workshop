```python
import numpy as np
```

```python
def getPowerSet(S):
    pS = set([])
    pS.add(frozenset([]))
    A = np.array(list(S))
    print(A)
    n = A.size
    for i in range(2**n):
        b = np.binary_repr(i,width=n)
        print(i)
        print(b)
        idx = np.array(list(b),dtype=int)==1
        print(A[idx])
        pS.add(frozenset(A[idx]))
    return pS
```

```python
S = set(np.arange(4))
pS = getPowerSet(S)
for i in pS:
    print(i)
```
